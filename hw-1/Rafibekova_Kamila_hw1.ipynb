```python
#используем One Hot кодирование категориальных переменных и нормализацию числовых
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler

CATEGORICAL = ['ninety_vintage', 'thirty_in_a_year', 'sixty_in_a_year', 'ninety_in_a_year', 'thirty_vintage', 'sixty_vintage', 'dlq_exist', 'if_zalog', 'INSURANCE_FLAG', 'SEX', 'EMPL_SIZE', 'INCOME_BASE_TYPE', 'EDUCATION', 'CREDIT_PURPOSE', 'EMPL_TYPE', 'EMPL_PROPERTY', 'EMPL_FORM', 'FAMILY_STATUS']
NUMERIC = ['DTI', 'FULL_AGE_CHILD_NUMBER', 'DEPENDANT_NUMBER', 'BANKACCOUNT_FLAG', 'Period_at_work', 'age', 'max90days', 'max60days', 'max30days', 'max21days', 'max14days', 'avg_num_delay', 'num_AccountActive180', 'num_AccountActive90', 'num_AccountActive60', 'Active_to_All_prc', 'numAccountActiveAll', 'numAccountClosed', 'sum_of_paym_months', 'all_credits', 'Active_not_cc', 'own_closed', 'min_MnthAfterLoan', 'max_MnthAfterLoan']

column_transformer = ColumnTransformer([
    ('ohe', OneHotEncoder(handle_unknown='ignore'), CATEGORICAL),
    ('scaling', StandardScaler(), NUMERIC)
])

transformed_df = column_transformer.fit_transform(df)

print(transformed_df.shape)
```

```python
#метод К-средних
from sklearn.cluster import KMeans

ks = []
for i in range(1, 10):
    kmeans = KMeans(n_clusters=i, random_state=0).fit(transformed_df)
    ks.append(kmeans.inertia_)

plt.plot(list(range(1, 10)), ks)
```

```python
#находим нужное число кластеров
kmeans = KMeans(n_clusters=5, random_state=0).fit(transformed_df)
unique, counts = np.unique(kmeans.labels_, return_counts=True)
dict(zip(unique, counts))
```

```python
first_class = [i for i, val in enumerate(kmeans.labels_) if val == 0]
second_class = [i for i, val in enumerate(kmeans.labels_) if val == 1]
third_class = [i for i, val in enumerate(kmeans.labels_) if val == 2]
forth_class = [i for i, val in enumerate(kmeans.labels_) if val == 3]
fifth_class = [i for i, val in enumerate(kmeans.labels_) if val == 4]
```

```python
for i, j in [
    (first_class, 'first'),
    (second_class, 'second'),
    (third_class, 'third'),
    (forth_class, 'forth'),
    (fifth_class, 'fifth')
]:
    df.iloc[i, :].to_csv(f'/Users/kamila/Desktop/{j}_class.csv')
```

```python
from string import ascii_letters
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="white")

# Compute the correlation matrix
corr = df.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
#RFM-analysys
recency = 'min_MnthAfterLoan'
frequency = 'numAccountActiveAll'
monetary = 'sum_of_paym_months'

for score_name, score in (
    ('RecencyScore', recency),
    ('FrequencyScore', frequency),
    ('MonetaryScore', monetary)
): 
    df[score_name] = pd.cut(
        df[score],
        bins=[
            -2,
            np.percentile(df[score], 20),
            np.percentile(df[score], 40),
            np.percentile(df[score], 65),
            np.percentile(df[score], 85),
            df[score].max()
        ],
        labels=[1, 2, 3, 4, 5]
    ).astype('int') 
    
df.loc[df.FrequencyScore == 1, 'FrequencyScore'] = 7
df.loc[df.FrequencyScore == 2, 'FrequencyScore'] = 6
df.loc[df.FrequencyScore == 4, 'FrequencyScore'] = 2
df.loc[df.FrequencyScore == 5, 'FrequencyScore'] = 1
df.loc[df.FrequencyScore == 7, 'FrequencyScore'] = 5
df.loc[df.FrequencyScore == 6, 'FrequencyScore'] = 4

df.loc[df.MonetaryScore == 1, 'MonetaryScore'] = 7
df.loc[df.MonetaryScore == 2, 'MonetaryScore'] = 6
df.loc[df.MonetaryScore == 4, 'MonetaryScore'] = 2
df.loc[df.MonetaryScore == 5, 'MonetaryScore'] = 1
df.loc[df.MonetaryScore == 7, 'MonetaryScore'] = 5
df.loc[df.MonetaryScore == 6, 'MonetaryScore'] = 4
```

```python
df['RFM'] = df.RecencyScore.astype(str) + df.FrequencyScore.astype(str) + df.MonetaryScore.astype(str)
```
